```python
# Render our plots inline
%matplotlib inline

import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
```

```python
projects = pd.read_csv("dubbo.csv")
```

```python
project_name="figures/dubbo"
```

```python
attributes = projects[['versions','Used Attributes Percentage']].copy()
```

```python
methods = projects[['versions','Used Methods Percentage']].copy()
```

```python
fig=plt.figure()
a=fig.add_axes((0.00,0.00,0.9,0.9)) # number here are coordinate (left,bottom,width,height)
b=fig.add_axes((1.0,0.00,0.9,0.9))

df1 = pd.DataFrame(attributes)
df2 = pd.DataFrame(methods)

df1.plot(x='versions',y='Used Attributes Percentage',rot=90,xticks=attributes.index,ax=a)
df2.plot(x='versions',y='Used Methods Percentage',rot=90,xticks=methods.index,ax=b)

plt.show()
fig.savefig(project_name+'/refused_bequest.png')
```
